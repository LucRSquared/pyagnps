```python
import sys
sys.path.append('..')

import matplotlib.pyplot as plt
from pyagnps.utils import polygonize_cell_reach_IDs_asc
```

```python
path_to_cell_IDs_asc = '../../inputs/topagnps/AnnAGNPS_Cell_IDs.asc' 
path_to_cell_IDs_shp = '../../outputs/shapefiles/AnnAGNPS_Cell_IDs.shp'

path_to_reach_IDs_asc = '../../inputs/topagnps/AnnAGNPS_Reach_IDs.asc' 
path_to_reach_IDs_shp = '../../outputs/shapefiles/AnnAGNPS_Reach_IDs.shp'
```

```python
gdf_cells = polygonize_cell_reach_IDs_asc(path_to_cell_IDs_asc, path_to_cell_IDs_shp, outepsg=4326, return_gdf=True)
gdf_reaches = polygonize_cell_reach_IDs_asc(path_to_reach_IDs_asc, path_to_reach_IDs_shp, outepsg=4326, return_gdf=True)
```

```python
fig, ax = plt.subplots(1, figsize=(20,15))

gdf_cells.plot(ax=ax, facecolor='none');
gdf_reaches.plot(ax=ax, column='DN', facecolor='none', linewidth=2);
```
