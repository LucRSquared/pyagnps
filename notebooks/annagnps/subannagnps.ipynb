```python
import matplotlib.pyplot as plt

from pyagnps import subannagnps
```

### AnnAGNPS files
- `AnnAGNPS_Cell_Data_Section.csv`
    Cells info file
- `AnnAGNPS_Reach_Data_Section.csv`
    Reaches info file
- `AnnAGNPS_Cell_IDs.asc`
    Raster file of cells IDs
- `AnnAGNPS_Reach_IDs.asc`
    Raster file of reach IDs

```python
inputs_folder = '../../../inputs/topagnps/'
outputs_folder = '../../../outputs/subannagnps/'

file_cell_section = inputs_folder + 'AnnAGNPS_Cell_Data_Section.csv'
file_reach_section = inputs_folder + 'AnnAGNPS_Reach_Data_Section.csv'
file_cell_raster = inputs_folder + 'AnnAGNPS_Cell_IDs.asc'
file_reach_raster = inputs_folder + 'AnnAGNPS_Reach_IDs.asc'
```

```python
df_cells = subannagnps.read_annagnps_section(file_cell_section)
df_cells
```

```python
df_reaches = subannagnps.read_annagnps_section(file_reach_section)
df_reaches
```

### Graph representation of the reaches network

```python
G = subannagnps.build_network(df_reaches)

plt.figure(figsize=(18,10))
subannagnps.draw_network(G)
plt.show()
```

### Outlet selection for a _sub_-network

```python
outlet = 31

subG = subannagnps.get_topagnps_subgraph(G, outlet=outlet)
plt.figure(figsize=(18,10))
subannagnps.draw_network(subG)
plt.show()
```

### Relabeled sub network
The last non-zero length reach is always labeled "2"

```python
topG, relabel_dict = subannagnps.relabel_topagnps_subgraph(subG, outlet=outlet)

plt.figure(figsize=(18,10))
subannagnps.draw_network(topG)
plt.show()
```

### Relabeling dictionary
The keys of the dictionary are the old reaches ids and the values are the new labels

```python
print(relabel_dict)
```

### Sub selection of AnnAGNPS data sections

```python
df_reaches_sub = subannagnps.subset_and_rename_reach_df(df_reaches, relabel_dict)
df_reaches_sub.head(10)
```

```python
df_cells_sub = subannagnps.subset_and_rename_cell_df(df_cells, relabel_dict)
df_cells_sub.head(10)
```

## Write new raster files

```python
outfile_cell_raster = outputs_folder+'AnnAGNPS_Cell_IDs.asc'
outfile_reach_raster = outputs_folder+'AnnAGNPS_Reach_IDs.asc'

subannagnps.subset_and_rename_cell_IDs_asc(df_cells_sub, file_cell_raster, outfile_cell_raster, keep_tmp_gtiff=False)
subannagnps.subset_and_rename_reach_IDs_asc(relabel_dict, file_reach_raster, outfile_reach_raster, keep_tmp_gtiff=False)
```

## Write shapefiles

```python
from src.pyagnps.utils import polygonize_cell_reach_IDs_asc

gdf_nested_cells = polygonize_cell_reach_IDs_asc(outfile_cell_raster, outfile_cell_raster.replace('.asc','_shp.shp'), outepsg=4326, return_gdf=True)
gdf_nested_reaches = polygonize_cell_reach_IDs_asc(outfile_reach_raster, outfile_reach_raster.replace('.asc','_shp.shp'), outepsg=4326, return_gdf=True)
```

## Write new data sections

```python
out_reach_csv = outputs_folder+'AnnAGNPS_Reach_Data_Section.csv'
out_cell_csv = outputs_folder+'AnnAGNPS_Cell_Data_Section.csv'

subannagnps.write_csv_data_section(df_reaches_sub, out_reach_csv)
subannagnps.write_csv_data_section(df_cells_sub, out_cell_csv)
```

## Visualization

```python
import matplotlib.pyplot as plt
import geopandas as gpd

CRS = 'epsg:4326'

gdf_container_cells = gpd.read_file('../../../outputs/shapefiles/AnnAGNPS_Cell_IDs.shp').to_crs(CRS)
gdf_container_reaches = gpd.read_file('../../../outputs/shapefiles/AnnAGNPS_Reach_IDs.shp').to_crs(CRS)

gdf_nested_cells = gdf_nested_cells.to_crs(CRS)
gdf_nested_reaches = gdf_nested_reaches.to_crs(CRS)
```

```python
fig, ax = plt.subplots(1, figsize=(25,15))

gdf_container_cells.plot(ax=ax, facecolor='tab:green', edgecolor='k', alpha=0.4);
gdf_nested_cells.plot(ax=ax, facecolor='tab:blue', alpha=0.5);
gdf_container_reaches.plot(ax=ax, facecolor='None', edgecolor='tab:green', linewidth=4, alpha=0.4);
gdf_nested_reaches.plot(ax=ax, facecolor='None', edgecolor='tab:blue', linewidth=3, alpha=0.8);
```
