# TopAGNPS outlet identification

- This notebook executes the commands to:
    - Run the preprocessing step of TopAGNPS from presribed boundary
    - Identify the outlet
    - Finish the computation

```python
import sys
sys.path.append('../..')
import geopandas as gpd
import matplotlib.pyplot as plt

from src.pyagnps import topagnps
from src.pyagnps import utils


import importlib
```

```python
files = {
    'uparea_asc': '../../../inputs/topagnps/topagnps_run1_run2/run1/UPAREA.ASC',
    'boundary': '../../../inputs/topagnps/topagnps_run1_run2/huc12_070802091104.gpkg'
}
```

```python
huc12 = gpd.read_file(files['boundary'])
huc12
```

```python
xout, yout, rowout, colout, raster_crs = topagnps.find_outlet_uparea_shape_intersection(files['uparea_asc'], huc12)
print(xout, yout, rowout, colout)
print(raster_crs)
```

```python
from shapely.geometry.point import Point
outlet = gpd.GeoDataFrame(geometry=[Point(xout, yout)], crs=raster_crs)
outlet = outlet.to_crs(huc12.crs)
```

```python
map = huc12.explore()
outlet.explore(m=map, color='red', marker_type='circle_marker', marker_kwds={'radius':8}, tooltip=False)
```
